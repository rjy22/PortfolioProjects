```python
# Guided project facilitated by Alex Freberg (YT: Alex the Analyst)
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create




# Read in the data

df = pd.read_csv(r'C:\Users\rjy22\OneDrive\Documents\Data Analytics Portfolio\AA_Guided Projects\movies.csv')
```

```python
# Looking at available data

df.head()
```

```python
# Looking for missing data

for col in df.columns:
    df = df.dropna()
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Looking at data types

df.dtypes
```

```python
# Changing data type of specific columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

```python
# Creating revised year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Checking for and dropping duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Investigating correlations
# Building scatter plot using matplotlib

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Film Budget')

plt.ylabel('Gross Earnings')

plt.show
```

```python
# Plotting budget vs gross using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Observing correlations

df.corr()
```

```python
# Visualizaing correlations

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Characteristics')

plt.xlabel('Movie Characteristics')

plt.ylabel('Movie Characteristics')

plt.show()
```

```python
# Looking at impact of company on correlation

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
# Sanity check

df.head()
```

```python
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Characteristics')

plt.xlabel('Movie Characteristics')

plt.ylabel('Movie Characteristics')

plt.show()
```

```python
df_numerized.corr()
```

```python
# Unstacking

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
